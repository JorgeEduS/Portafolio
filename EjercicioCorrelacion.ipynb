```python
# Incluya aquí su código para importar librerías y datos
import numpy as np
import pyspark
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("responses.csv")
df.head()
```

```python
df.shape
```

```python
df.head()
```

```python
# Incluya aquí el código para obtener las correlaciones
matriz_correlacion = df.corr()
matriz_correlacion

# Aqui podemos observar la matriz de correlaciones entre las columnas, podemos observar que que tenemos valores bastante bajos, tanto negativos como positivos. Podemos observar los casos de
# Musica con slow songs/fast songs, los cuales cuentan con una correlacion positiva sin embargo, baja ya que es cercana a cero. 
# Podemos observar una correlacion media con valor de 0.386646 entre Country-Folky Musica clasica-folk.
```

```python
plt.matshow(matriz_correlacion)
# Aqui observamos la matriz de correlacion en un heatmap, observando como la linea se distribuye considerando la correlacion maxima de 1.0
```

```python
f, ax = plt.subplots(figsize=(10,8))
sns.heatmap(matriz_correlacion, mask=np.zeros_like(matriz_correlacion, dtype = np.bool), cmap = sns.diverging_palette(220, 10, as_cmap = True), square=True, ax=ax)
# Aqui podemos observar nuevamente lo que vimos en el anterior heatmap, sin embargo, podemos observar ahora los nombres para cada valor. 
# Podemos observar como muchos valores se encuentran dentro de la zona de -0.2 a 0.2, siendo los mas predominantes, con algunos pocos que llegan hasta -0.4 y otros que llegan a 0.6-0.8.
```

```python
# Código para crear un dataframe solo con los campos de variables Música y Demografía
variables_musica = list(df.columns[0:10])
variables_demografía = list(df.columns[-10:])
lista_variables = variables_musica + variables_demografía
lista_variables
```

```python
data_mus_dem = df[lista_variables]
data_mus_dem
```

```python
# Incluye aquí tu código para hacer el análisis de correlaciones
correlacion_mus_dem = data_mus_dem.corr()
correlacion_mus_dem

# Podemos observar la correlacion nuevamente pero de las columnas previamente establecidas. 
# Podemos ver que hay correlaciones mas fuertes en casos como el Rock-Metal, al igual que como era de esperarse, entre el Altura-Peso. 
# Sin embargo, no encontramos muchsa correlaciones fuertes entre los valores. 
```

```python
# Ejemplo: Matriz de correlación de variables de música y demografía
f,ax = plt.subplots(figsize=(10, 8)) 
corr = data_mus_dem.corr() 
sns.heatmap(corr, mask=np.zeros_like(corr, dtype=np.bool), cmap=sns.diverging_palette(220, 10, as_cmap=True), 
      square=True, annot=True, ax=ax) 

# Nuevamente podemos ver un heatmap con los valores de la correlacion, podemos observar caso como el Dance-Pop, Rock-Metal, Weight-Height resaltar 
# con correlaciones altas, que fueron los que previamente mencionamos.
# Y con otros tantos como Metal-Dance, o Musical-Height, los cuales tienen una correlacion negativa considerable.
```
