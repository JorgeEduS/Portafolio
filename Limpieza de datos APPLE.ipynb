```python
# Aqui guardamos la combinacion de todos los valores en una nueva variable, llamada toda_venta_df, donde podemos visualizar que ahora llegamos hasta los 186849 registros, a comparacion de enero
# o febrero, que podemos visualizar que unicamente llegan a 9722/12035 registros, por lo que podemos establecer que se guardaron de manera correcta.
toda_venta_df = pd.concat((pd.read_csv(file)for file in archivos_venta), ignore_index=True)
toda_venta_df
```

```python
# Tendremos que realizar una limpieza de datos, como vimos anteriormente, en el valor numero 2, contamos con una fila NaN. Asi que queremos observar cuanta cantidad
# de estos mismos tenemos dentro de nuestro dataset.
toda_venta_df.isnull().sum()

# Como podemos observar, en el dataframe que tenemos, contamos con al rededor de 545 datos nulos/NaN, los cuales tendremos que realizarle una limpieza. 
```

```python
# Como pudimos observar la primera vez que imprimimos el dataset, pudimos ver que no solamente era un valor en la columna, si no, era toda la fila con valores NaN. 
# Esto es algo importante ya que asi podemos saber como manejar el NaN, ya que en caso de que solo fuera un valor o dos por fila, no podriamos solamente borrarlos.
valores_NaN = toda_venta_df[toda_venta_df.isna().any(axis=1)]
valores_NaN

# Nuevamente, podemos observar lo que se mencionaba anteriormente, pero con diferentes filas, de las cuales cada una de ellas esta completamente en NaN.
```

```python
# Aqui lo que hacemos es eliminar toda fila que cuente con un NaN, de esta forma ya no deberiamos tener valores nulos dentro de nuestro dataset.
toda_venta_df = toda_venta_df.dropna(how = "all")
toda_venta_df
```

```python
# Aqui creamos este metodo, el cual usaremos para obtener y separar la ciudad del dataframe que tenemos. 
def obtenerCiudad(x):
    # Aqui utilizamos el separador, ya que como vemos en la columna Purchase Adress tenemos la calle, ciudad y numero.
    # lo que hacemos es que guardamos la direccion en diferentes partes, separandolas por la columna. 
    direccionSeparada = x.split(',')
    # Aqui lo que hacemos es que revisamos si la direccion cuenta con mas de dos partes, ya que anteriormente no me quiso funcionar debido a que se salia del index. 
    # con este metodo, verifica que si exista ese index y en caso de que si, lo agrega y corta dejando unicamente la ciudad.
    if len(direccionSeparada) >= 2:
        return direccionSeparada[1].strip()
    # En caso de que no, unicamente dejamos la ciudad. 
    else:
        return ""
# Aqui finalmente, creamos una columna nueva para ciudad, en base a Purchase Adress y asi le damos la ciudad respectiva.
toda_venta_df['City'] = toda_venta_df['Purchase Address'].apply(obtenerCiudad)
toda_venta_df
```

```python
# Repetimos el metodo anterior, pero para poder obtener el valor de el Dia.
def obtenerDia(x):
    diaSeparado = x.split('/')
    if len(diaSeparado) >= 2:
        return diaSeparado[1].strip()
    else:
        return ""
toda_venta_df['Day'] = toda_venta_df['Order Date'].apply(obtenerDia)
toda_venta_df
```

```python
# Repetimos el metodo anterior, pero para poder obtener el valor de el Mes.
def obtenerMes(x):
    mesSeparado = x.split('/')
    if len(mesSeparado) >= 2:
        return mesSeparado[0].strip()
    else:
        return ""
toda_venta_df['Month'] = toda_venta_df['Order Date'].apply(obtenerMes)
toda_venta_df
```
